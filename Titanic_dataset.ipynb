```python
import pandas as pd
import numpy as np

data = pd.read_csv("/content/drive/MyDrive/PROJECT/Titanic-Dataset.csv")
print(data.head())
```

```python
age_series = data['Age']
fare_series = data['Fare']
survived_series = data['Survived']

print(age_series.head())
print(fare_series.head())
print(survived_series.head())
```

```python
print(data.head())
print(data.info())
print(data.describe())
```

```python
print(data.isnull().sum())
data['Age'].fillna(data['Age'].mean(), inplace=True)

data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)

data.drop(columns=['Cabin'], inplace=True)
print(data.isnull().sum())
```

```python
print(data.loc[0])
print(data.iloc[0:5])
print(data[['Age', 'Fare', 'Survived']].head())
```

```python
data.rename(columns={'Sex': 'Gender'}, inplace=True)

print(data.head())
pivot_table = data.pivot_table(values='Survived',
                             index='Gender',
                             aggfunc='sum')

print(pivot_table)
```

```python
age_above_30 = data[data['Age'] > 30]
fare_above_50 = data[data['Fare'] > 50]
female_survived = data[(data['Gender'] == 'female') & (data['Survived'] == 1)]
print(female_survived.head())
```

```python
print(data.sort_values(by='Fare', ascending=False).head())
print(data.sort_values(by='Age').head())
print(data.sort_values(by='Survived', ascending=False).head())
```

```python
survival_by_gender = data.groupby('Gender')['Survived'].mean()
print("Survival Rate by Gender:")
print(survival_by_gender)
```

```python
avg_age_by_class = data.groupby('Pclass')['Age'].mean()
print("Average Age by Passenger Class:")
print(avg_age_by_class)
```

```python
survival_by_embarked = data.groupby('Embarked')['Survived'].sum()
print("Survival Count by Embarked:")
print(survival_by_embarked)
```

```python
category_df = pd.DataFrame({
    'Pclass': [1, 2, 3],
    'Class_Category': ['Upper', 'Middle', 'Lower']
})
merged_data = pd.merge(data, category_df, on='Pclass')
print(merged_data.head())
```

```python
data_join = data.join(category_df.set_index('Pclass'), on='Pclass')
print(data_join.head())
```

```python
data_part1 = data.iloc[:200]
data_part2 = data.iloc[200:400]

concat_data = pd.concat([data_part1, data_part2])
print(concat_data.head())
```

```python
avg_age_survival = data.groupby('Survived')['Age'].mean()
print(avg_age_survival)
```

```python
print("Highest Fare:", data['Fare'].max())
print("Lowest Fare:", data['Fare'].min())
```

```python
survival_percentage = (data['Survived'].sum() / len(data)) * 100
print("Overall Survival Percentage:", survival_percentage)
```
